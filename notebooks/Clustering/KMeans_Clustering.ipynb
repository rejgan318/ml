## Алгоритм K-Means

1. Первым шагом в k-средних является выбор количества кластеров, k.
2. Выберите k случайных точек из данных в качестве центроидов
3. Перебираем все точки и приписываем их к кластеру, у которого ближайший текущий центроид
4. Вычисляем новые центроиды для новых кластеров
5. Повторяем 3 и 4 до тех пор, пока не исчерпается заданное количество итераций либо центроиды перестанут меняться или точки остаются в одном кластере

Выбор начальных кластеров можно делать не случайно, а используя метод K-Means++ (специальный алгоритм, первый кластер случайный, следующие начальные вычисляются)

Дальше говнокод реализации метода

```python
# Step 3 - Assign all the points to the closest cluster centroid
# Step 4 - Recompute centroids of newly formed clusters
# Step 5 - Repeat step 3 and 4

diff = 1
j=0

while (diff != 0):
    XD=X
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["ApplicantIncome"]-row_d["ApplicantIncome"])**2
            d2=(row_c["LoanAmount"]-row_d["LoanAmount"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        X[i]=ED
        i=i+1

    C=[]
    for index,row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["LoanAmount","ApplicantIncome"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['LoanAmount'] - Centroids['LoanAmount']).sum() + (Centroids_new['ApplicantIncome'] - Centroids['ApplicantIncome']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["LoanAmount","ApplicantIncome"]]
```

## Как выбрать правильное количество кластеров в кластеризации K-средних?

Максимально возможное количество кластеров будет равно количеству наблюдений в наборе данных. Но как тогда определить оптимальное количество кластеров? Одна вещь, которую мы можем сделать, это построить график, также известный как кривая локтя, где ось x будет представлять количество кластеров, а ось y будет метрикой оценки. Скажем, инерция или индекс Данна. значение кластера, при котором это уменьшение значения инерции становится постоянным, может быть выбрано в качестве правильного значения кластера для наших данных .

## Реализация кластеризации K-средних в Python

Цель этой задачи состоит в том, чтобы сегментировать клиентов оптового дистрибьютора на основе их годовых расходов на различные категории продуктов, такие как молоко, бакалея, регион и т. д.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans

# reading the data and looking at the first five rows of the data
data=pd.read_csv("KMeans_Clustering_Wholesale_customers_data.csv")
data.head()
```

У нас есть информация о расходах клиентов на различные продукты, такие как молоко, бакалея, замороженные продукты, моющие средства и т. д. Теперь нам нужно сегментировать клиентов на основе предоставленных данных. Прежде чем сделать это, давайте вытащим некоторую статистику, связанную с данными:

```python
# statistics of the data
data.describe()
```

Здесь мы видим, что существует много различий в величине данных. Такие переменные, как «Канал» и «Регион», имеют низкую величину, тогда как такие переменные, как «Свежесть», «Молоко», «Бакалея» и т. д., имеют более высокую величину.

Поскольку K-Means — это алгоритм, основанный на расстоянии, эта разница величин может создать проблему. Итак, давайте сначала приведем все переменные к одинаковой величине, т.е. пронормируем:

```python
# standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# statistics of scaled data
pd.DataFrame(data_scaled).describe()
```

```python
# defining the kmeans function with initialization as k-means++
kmeans = KMeans(n_clusters=2, init='k-means++')

# fitting the k means algorithm on scaled data
kmeans.fit(data_scaled)
```

Мы инициализировали два кластера и обратите внимание — здесь инициализация не случайна. Мы использовали инициализацию k-means++, которая обычно дает лучшие результаты, как мы обсуждали в предыдущем разделе.

Оценим, насколько хорошо сформированы кластеры. Для этого рассчитаем инерцию кластеров:

```python
# inertia on the fitted data
kmeans.inertia_
```

Мы получили значение инерции почти 2600. Теперь давайте посмотрим, как мы можем использовать кривую локтя для определения оптимального количества кластеров в Python.

Сначала мы подберем несколько моделей k-средних, и в каждой последующей модели мы будем увеличивать количество кластеров. Мы сохраним значение инерции каждой модели, а затем построим график, чтобы визуализировать результат (Кривая локтя). Искомое количество кластеров будет лежать в области, где быстрее всего меняется значение критерия оценки (здесь - инерции)

```python
# fitting multiple k-means algorithms and storing the values in an empty list
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters = cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Можете ли вы определить оптимальное значение кластера по этому графику? Глядя на приведенную выше кривую локтя, мы можем выбрать любое количество кластеров от 5 до 8 . Давайте установим количество кластеров равным 6 и подгоним модель:

```python
# k means using 5 clusters and k-means++ initialization
kmeans = KMeans(n_clusters = 5, init='k-means++')
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)
```

Наконец, давайте посмотрим на количество баллов в каждом из сформированных выше кластеров:

```python
frame = pd.DataFrame(data_scaled)
frame['cluster'] = pred
frame['cluster'].value_counts()
```
